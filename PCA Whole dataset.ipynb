This notebook computes and analizes the PCA of the mean data of the MEG. 

```r
library(tidyverse)


meg_data <- readRDS("/home/rstudio/data/meg_mci.rds")
meg_data_nums <- meg_data %>% select(-id, -class)
```

Compute the PCA of the mean data, without taking into account the ID and Class columns.

```r
meg_pca <- prcomp(meg_data_nums, scale = TRUE)
meg_pca_predict <- predict(meg_pca)
```

```r
library(ggfortify)
autoplot(meg_pca, data = meg_data, colour = 'class')
```

```r
set.seed(4)
autoplot(kmeans(meg_pca_predict, 2), data = meg_pca_predict)
```

```r
library(cluster)
```

```r
autoplot(clara(meg_pca_predict, 2), scale = TRUE)
```
