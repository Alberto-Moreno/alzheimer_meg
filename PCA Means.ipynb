This notebook computes and analizes the PCA of the mean data of the MEG. 

```r
library(tidyverse)


mean_data <- readRDS("/home/rstudio/data/mean.rds")
mean_data_nums <- mean_data %>% select(-id, -class)
```

Compute the PCA of the mean data, without taking into account the ID and Class columns.

```r
mean_meg_pca <- prcomp(select(mean_data, -id, -class), scale = TRUE)
mean_pca_predict <- predict(mean_meg_pca)
```

```r
library(ggfortify)
autoplot(mean_meg_pca, data = mean_data, colour = 'class')
```

```r
class(mean_pca_predict)
```

```r
set.seed(4)
autoplot(kmeans(mean_pca_predict, 2), data = mean_pca_predict)
```
